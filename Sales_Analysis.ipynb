# **DATA VISUALIZATION**



```python
#count of each sub-category using sns.countplot
plt.figure(figsize = (24,6))
sns.countplot('Sub-Category',data = df)
```

```python
#count of each category without using sns.countplot
data = df.groupby(['Category']).agg({'Sales':['count']})
data.columns = ['No of products purchased']
data = data.reset_index()
dff = pd.DataFrame(data)
dff.head()
```

```python
dff = dff.head()
d = pd.DataFrame(dff , columns=["Category","No of products purchased"])
d.plot(x="Category",y="No of products purchased",kind="bar",figsize = (10,10))
d.plot(x="Category",y="No of products purchased",kind="line",figsize = (10,10))
```

```python
#sub-category vs sales
d1 = pd.DataFrame(df , columns=["Sub-Category","Sales"])
d1.plot(x="Sub-Category",y="Sales",kind="line",figsize = (40,10))           
```

```python
#sales of the month based on the order date
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = (df["Order Date"]).dt.month
plt.figure(figsize = (15,6))
plt.style.use('classic') 
SalesTime = sns.lineplot(x = "Month" , y = "Profit" , data = df)
SalesTime.set_title(label = "Product sales per month" , fontsize = 25 , color="Red")
```

```python
df.corr
```

```python
plt.figure(figsize = (10,10))
hm = sns.heatmap(df.corr() , annot = True , cmap = "icefire")
hm.set_title(label = "Correlation Matrix" , fontsize = 25 , color="Red")
```

```python
#average sales of different sub-category 
plt.figure(figsize = (15,6))
plt.style.use('classic') 
ax = sns.boxenplot(x = "Quantity" , y = "Sub-Category" ,data = df)
ax.set_title(label = "Average sales of different sub-category products" , fontsize = 25 , color="Red")
ax.set_xlabel(xlabel = "Quantity" , fontsize = 15)
ax.set_ylabel(ylabel = "Sub-Category products" ,fontsize = 15)
```

```python
#creating dummies for segment column 
dummies = pd.get_dummies(df['Segment'])
print(dummies)
```

```python
df = pd.concat([df,dummies],axis = 1)
df.iloc[37:50]
```

```python
sns.lineplot(x = 'Category', y = 'Home Office',data = df)
```

```python
#plot representing the product that is purchased by many people
sns.countplot(df['Category'] , palette = 'crest')
plt.xlabel('Product mostly sold')
plt.ylabel('Count')
plt.title('Which Product is mostly sold?' , color = "Red" , fontsize = 25)
plt.show()
```

```python
#pie chart representing the region that mostly purchases the products
def pie_chart(df):
    labels = df.index.values
    sizes = df['Region'].values
    fig , ax = plt.subplots()
    ax.pie(sizes , labels = labels , autopct = "%10.1f%%" , shadow = True , startangle = 90 ,colors = ['#DF962F','#66b3ff','#c2c2f0','#ffb3e6'] )
    ax.set_title("Which region people mostly purchased?" , fontsize = 25 , color = "Red")
    ax.axis('equal')
    plt.show()
sf = pd.DataFrame(df.Region.value_counts())
pie_chart(sf)
```

```python
plt.figure(figsize = (15,15))
plt.style.use('classic')
ax = sns.relplot(x= "Profit" , y = "Sub-Category" , data = df)
```

```python
#product sales based on category
plt.style.use('classic')
plt.figure(figsize = (14,6))
ax= sns.stripplot(y= "Sub-Category", x = "Sales", hue = "Sub-Category", data = df)
ax.set_title(label = "Product sales on the basis of Category" , fontsize = 25 ,color = "Red")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Sub-Category")
```

```python
#histogram for integer and float datatype columns after dropping row ID
df=df.drop(['Row ID'] , axis = 1)
sa = df.iloc[:]
sa.hist(figsize = (12,12))
plt.show()
```

```python
#distplot representing quantity
plt.figure(figsize = (18,6))
sns.distplot(df["Quantity"])
```

```python
#pairplot representing sub-category
sns.pairplot(df , hue = "Sub-Category")
plt.show()
```

```python
#scatterplot representing sales vs profit
plt.figure(figsize = (15,6))
ax = sns.scatterplot(x = df['Sales'] , y = df['Profit'] , hue = df['Segment'] , size = df['Sales'] , palette = "icefire" , legend = "auto" , sizes =(100,500))
ax.set_title("Sales vs Profit" , fontsize = 25 , color = "Red")
plt.show()
```
